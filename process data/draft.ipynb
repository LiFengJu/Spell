```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('restaurant_data1.csv')

```

```python
df.iloc[:, 1] = df.iloc[:, 1].str[18:]
df.iloc[:, 3] = df.iloc[:, 3].str[:-3]
df.iloc[:, 2] = df.iloc[:, 2].astype(str).apply(lambda x: x.rsplit('¥', 1)[-1])
```

```python
from bs4 import BeautifulSoup

def extract_rating_divs(html_string):
    soup = BeautifulSoup(html_string, 'html.parser')

    rating_divs = soup.find_all('div', class_='DzMcu')

    return rating_divs

```

```python
def parse_rating_divs(rating_divs):
    parsed_results = []
    for div in rating_divs:
        rating_name = div.find('span', class_='BPsyj').text
        rating_span = div.find('span', class_='ui_bubble_rating')
        if rating_span:
            rating = int(rating_span['class'][1].split('_')[-1]) / 10
            parsed_results.append((rating_name, rating))
        else:
            parsed_results.append((rating_name, None))
    return parsed_results

df['新列'] = df['待处理1'].apply(extract_rating_divs).apply(parse_rating_divs)
for i, row in df.iterrows():
    for rating_name, rating_value in row['新列']:
        df.loc[i, rating_name] = rating_value
```

```python
df = df.drop(['待处理1', '待处理2', '新列'], axis=1)
df
```

```python
df_cleaned = df.dropna()
```
